```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
siny = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
```

```python
plt.plot(x,siny,x,cosy,color="red",linestyle="--")
plt.xlim(-5,15)
plt.ylim(-2,2)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.axis([-2,12,-2,2])
```

```python
plt.plot(x,siny)
plt.plot(x,cosy)
plt.xlabel("x axis")
plt.ylabel("y value")
```

```python
plt.plot(x,siny,label="sin(x)",color="r",linestyle="-")
plt.plot(x,cosy,label="cos(x)",color="k",linestyle="--")
plt.axis([-2,12,-2,2])
plt.legend(loc="best")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.title("title")
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy)
```

```python
plt.scatter(x,siny,s=x**2,c=x)
```

```python
plt.scatter(x,siny,s=siny**2*100,c=siny)
plt.scatter(x,cosy,s=cosy**2*100,c=cosy)
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.scatter(x,y,s=np.abs(x+y)**5,c=x+y,alpha=0.5)
```

```python
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.scatter(x,y,s=(np.abs(x)+np.abs(y))**3,c=(np.abs(x)+np.abs(y)),alpha=0.5)
```
